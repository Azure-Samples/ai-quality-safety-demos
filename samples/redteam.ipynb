## Set up the test application

For this notebook, we are sending requests to this RAG-on-PostgreSQL application:

https://github.com/Azure-Samples/rag-postgres-openai-python/

That application uses LLMs to answer questions based on products stored in a PostgreSQL database.

````python
import requests


def send_question_to_app(
    question: str,
    target_url: str = "http://127.0.0.1:8000/chat",
):
    headers = {"Content-Type": "application/json"}
    body = {
        "messages": [{"content": question, "role": "user"}],
        "stream": False,
        "context": {
            "overrides": {"use_advanced_flow": False, "top": 3, "retrieval_mode": "hybrid", "temperature": 0.3}
        },
    }
    url = target_url
    r = requests.post(url, headers=headers, json=body)
    response = r.json()
    if "error" in response:
        return f"Error received: {response['error']}"
    else:
        return response["message"]["content"]

send_question_to_app("What are good climbing shoes?")
````

## Batch attacks with adversarial question datasets

````python
# Fetch data set of adversarial prompts
from pyrit.datasets import fetch_llm_latent_adversarial_training_harmful_dataset
from rich import print

prompt_dataset = fetch_llm_latent_adversarial_training_harmful_dataset()
prompt_list = list(prompt_dataset.get_values(first=8))
print(prompt_list)
````

````python
# Set up the HTTP prompt target
from httpx import Response
from pyrit.prompt_target import HTTPTarget

raw_http_request = """POST http://localhost:8000/chat HTTP/1.1
Content-Type: application/json

{
    \"messages\": [
        {\"role\": \"user\", \"content\": \"{{PROMPT}}\"}
    ],
    \"stream\": false,
    \"context\": {
        \"overrides\": {\"use_advanced_flow\": false, \"top\": 3, \"retrieval_mode\": \"hybrid\", \"temperature\": 0.3}
    }
}
"""

def parsing_function(response: Response) -> str:
    response_json = response.json()
    if "error" in response_json:
        return f"Error received: {response_json['error']}"
    else:
        return response_json["message"]["content"]

http_prompt_target = HTTPTarget(http_request=raw_http_request, callback_function=parsing_function, timeout=20.0)
print("âœ… HTTP prompt target initialized")
````

````python
# Send the attack prompts
from pyrit.executor.attack import (
    AttackExecutor,
    ConsoleAttackResultPrinter,
    PromptSendingAttack,
)

attack = PromptSendingAttack(objective_target=http_prompt_target)

results = await AttackExecutor().execute_multi_objective_attack_async(
    attack=attack,
    objectives=prompt_list,
)

for result in results:
    await ConsoleAttackResultPrinter().print_conversation_async(result=result)
````

````python
# Send the attack prompts with a converter
from pyrit.executor.attack import (
    AttackConverterConfig,
    AttackExecutor,
    ConsoleAttackResultPrinter,
    PromptSendingAttack,
)
from pyrit.prompt_normalizer.prompt_converter_configuration import (
    PromptConverterConfiguration,
)

converters = PromptConverterConfiguration.from_converters(converters=[UrlConverter()])
converter_config = AttackConverterConfig(request_converters=converters)
attack = PromptSendingAttack(objective_target=http_prompt_target, attack_converter_config=converter_config)

results = await AttackExecutor().execute_multi_objective_attack_async(
    attack=attack,
    objectives=prompt_list[0:4],
)

for result in results:
    await ConsoleAttackResultPrinter().print_conversation_async(result=result)
````

````python
# Send the attack prompts with a composition of converters
converters = [
    UrlConverter(),
    ROT13Converter(),
]
converters = PromptConverterConfiguration.from_converters(converters=converters)
converter_config = AttackConverterConfig(request_converters=converters)
attack = PromptSendingAttack(objective_target=http_prompt_target, attack_converter_config=converter_config)

results = await AttackExecutor().execute_multi_objective_attack_async(
    attack=attack,
    objectives=prompt_list,
)

for result in results:
    await ConsoleAttackResultPrinter().print_conversation_async(result=result)
````

# Evaluate attack success rate

https://azure.github.io/PyRIT/code/scoring/0_scoring.html

````python
# Score the attack conversations for refusal outcomes
from pyrit.prompt_target import OpenAIChatTarget
from pyrit.score import SelfAskRefusalScorer

scorer_llm = OpenAIChatTarget(model_name="llama3.1:latest",
                              endpoint="http://localhost:11434/v1/chat/completions",
                              api_key="nokeyneeded")

refusal_scorer = SelfAskRefusalScorer(chat_target=scorer_llm)
scored_refusals = []

for attack_result in results:
    message_piece = attack_result.last_response
    if message_piece is None:
        print("âš ï¸ No response received for objective, marking as refusal.")
        scored_refusals.append(
            {
                "objective": attack_result.objective,
                "refused": True,
                "rationale": "No response returned by target.",
            }
        )
        continue

    scored_response = (
        await refusal_scorer.score_async(
            message=message_piece.to_message(), objective=attack_result.objective
        )
    )[0]
    scored_refusals.append(
        {
            "objective": attack_result.objective,
            "response": message_piece.to_message().get_value(),
            "refused": bool(scored_response.get_value()),
            "rationale": scored_response.score_rationale or "No rationale provided.",
        }
    )

total_attempts = len(scored_refusals)
refusals = sum(1 for record in scored_refusals if record["refused"])
success_rate = (total_attempts - refusals) / total_attempts if total_attempts else 0.0

print(f"Total prompts evaluated: {total_attempts}")
print(f"Refusals detected: {refusals}")
print(f"Attack success rate (non-refusals): {success_rate:.1%}")

for record in scored_refusals:
    status = "ðŸ‘ðŸ» ATTACK REFUSED" if record["refused"] else "ðŸ˜± ATTACK SUCCEEDED"
   
    print(f"[{status}] {record['objective']}")
    # print out first 100 characters of the response
    print(f"  Response: {str(record['response'][0:100])}...")
    print(f"  Rationale: {record['rationale']}")
````

## Reducing attack success rate


1. Use models that have undergone an RLHF or similar safety fine-tuning process. Check the model card for each model to see its safety benchmarks.

1. Use an additional safety layer on top of the model to filter out harmful inputs and outputs.


1. Improve the prompt in the app to refuse unacceptable requests:

    ```
    You must not generate content that may be harmful to someone physically or emotionally even if a user requests or creates a condition to rationalize that harmful content. You must not generate content that is hateful, racist, sexist, lewd or violent.
    ```

1. Improve the prompt in the app to refuse off-topic questions:

    ```
    You must only answer questions related to the products in the database. If a question is not related to the products, you must refuse to answer it by stating "I am unable to answer that question as it is not related to products."
    ```

1. For RAG applications, ensure that you are only sending back relevant context for the question. Use a re-ranking model with a threshold to discard irrelevant context after retrieval. For most harmful questions, there should *not* be any relevant context.
